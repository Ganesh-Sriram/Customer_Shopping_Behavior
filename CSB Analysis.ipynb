```python
import pandas as pd
```

```python
df = pd.read_csv("customer_shopping_behavior.csv")

display(df.head())
```

```python
df.info()
```

```python
df.isnull().sum()
```

## Null Handling

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
display(df.groupby('Category')['Review Rating'].mean())
```

## Conversion of ColumnNames suitable for Scalable Usage

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.info()
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.info()
```

## Creating New Columns for Better Data Underestanding

```python
# Creating column for Age Group

labels = ["Young Adult", "Adult", "Middle-aged", "Senior"]

df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
# Creating column for Purchase Frequency Days

frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Annually' : 365,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Monthly': 30,
    'Every 3 Months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df = df.drop('promo_code_used', axis = 1)
```

```python
df.head()
```

## Saving the File

```python
df.to_csv('csb_cleaned.csv', index=False)
```
